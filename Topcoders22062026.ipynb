# TypeCasting

We know python dynamically assign the data type to the variable. After the variable is assigned with the data can we change that ? Yes using few built in function we can change the data type.

```python
# When a string literal is assigned to variable, the variable becomes string variable
quantity = "5"
price = "1455.25"
print(quantity, type(quantity), price, type(price))
# now we cannot perform arithmatic operation like multiplication
# To perform multiplication, We must change the data type from string to integer or float
quantity = int(quantity) # typecasted to integer using int function
price = float(price) # typecasted to float using float function
print(quantity, type(quantity), price, type(price))
print("Total amount to be paid", quantity * price)
```

```python
carplate = "JCG"
carplatenumber = 9876
print(carplate, type(carplate), carplatenumber, type(carplatenumber))
# in this case we cannot typecast "JCG" to a number
# so we typecast 9876 to string and then perform string concatenation
carplatenumber = str(carplatenumber) # convert anything to string we can use str function
print(carplate + carplatenumber) # adding string to another string is called string concatenation
```

```python
# Let us talk about boolean
# If a variable hold a value other than 0, typecast it to boolean you will get True
# If a variable hold a value equals to 0, typecast it to boolean you will get False
x = 2823
print(bool(x)) # True
x = 0
print(bool(x)) # False
```

```python
# If a variable hold a value other than "", typecast it to boolean you will get True
# If a variable hold a value equals to "", typecast it to boolean you will get False
x = "Hello"
print(bool(x)) # True
x = ""
print(bool(x)) # False
```

# List

In Python language we have a data structure called List to hold more than a single value and assign it to a single variable. In Python we do not use the word array instead we use the word List.

```python
# multiple values are seperated by comma and placed inside a box []
# the box is assigned to the variable fruits
fruits = ["apple", "orange", "mango", "banana", "grapes"]
print(fruits)
print(type(fruits))
```

# Selection and Indexing

```python
print(fruits)
print(fruits[0])
print(fruits[1])
print(fruits[2])
print(fruits[3])
print(fruits[4])
```

```python
# How can i find the number of items in the list
# There is a built in function called len
# The function len tell us how many items we have in an iterable objects
print(len(fruits))
```

```python
# You want to pull out the last item in the list
print(fruits[len(fruits) - 1]) # print(fruits[5 - 1]) print(fruits[4])
print(fruits[len(fruits) - 2])
print(fruits[len(fruits) - 3])
```

```python
# In python the index number can be negative
print(fruits[-1])
print(fruits[-2])
print(fruits[-3])
print(fruits[-4])
print(fruits[-5])
```

```python
# In python the index can be range
# Create a slice of the original list using range
print(fruits)
print(fruits[1:3]) # start index is included and end index is not included
# in the above example item at index 1 is included, item at index 2 is included
# however item at index 3 is not included
# If you want banana also then you must mention 4 instead of 3
print(fruits[1:4]) # range has start index and end index seperated by : (colon)
```

```python
# What if I did not mention the start index
print(fruits[:4]) # since start index is not mention it starts from 0 => 0, 1, 2, 3 (4 is not included)
```

```python
# What if I did not mention the end index
print(fruits[2:]) # since end index is not mention it goes all the way to the last index
# => 2, 3, 4
```

```python
print(fruits[:]) # same as print(fruits)
```

```python
# the range can have 3rd parameter which is step up value (increment value)
print(fruits[::2]) # increment the iteration by 2
# meaning item at index 0, 2, 4 is retrieved
```

```python
# In a range the start index is suppose to be smaller than the end index.
# However if the increment value is negative then start index can be greater than end index
print(fruits[4:0:-1])
```

1. len(fruits) => function call

# Tuple

Tuple is nothing but readonly list

```python
# Instead of using square bracket we use round bracket () to create a tuple
# Remember tuple is nothing but readonly list
# Meaning selection and indexing will be exactly same as list
# However methods like append, remove, pop are not available in the list
countries = ("Indonesia", "Malaysia", "Singapore", "Thailand")
print(countries)
print(countries[0])
print(countries[1])
print(countries[2])
print(countries[3])
print(len(countries))
print(countries[-1])
print(countries[-2])
print(countries[-3])
# countries.append("Vietnam") # this will throw an error 'tuple' object has no attribute 'append'
# Remember it is readonly list
```

```python
# Since tuple is readonly, performance wise it is faster than a list
# Tuple is a preferred data structure for the python language.
prices = 10.5, 20.5, 30.5, 40.5, 50.5 # Here we did not use () or [] Values are just seperated by comma
# and assigned to the variable prices.
# Python will automatically place the values inside the tuple and assign the tuple to the variable prices
print(prices, type(prices))
```

The () is used for the following purposes
1. call a function ex. print()
2. evaluate expression ex. (10 + 5) * 3
3. to create a tuple ex. ("apple", "orange", "mango")


```python
# Problem is if you have a single item in the tuple it will automatically become an integer
mynumbers = (10) # tuple with single item
print(mynumbers, type(mynumbers))
# How can i create a tuple with single item ?
mynumbers = (10,) # inside an expression you will not have comma so python will treat this as tuple
print(mynumbers, type(mynumbers))
```

# Unpacking

The iterable objects like list, tuple gets unpacked automatically.

```python
fruits = ["apple", "orange", "mango"]
print(fruits)
# Only in python on the left hand side of the statement you can have more than a single variable
# When you have more than 1 variable on the left hand side python will automatically look for an
# iterable object on the right hand side of the statement.
fruitone, fruittwo, fruitthree = fruits # fruits must be an iterable oject
print(fruitone)
print(fruittwo)
print(fruitthree)
```

# Arithmatic Operators
They always return numeric values

```python
x = 10
y = 5
z = 3
print(x + y) # addition
print(x - y) # subtraction
print(x * y) # multiplication
print(x / z) # division
print(x // z) # integer division (quotient)
print(x % z) # remainder
print(x ** z) # exponentials 10 x 10 x 10
```

# Comparision Operators
Always returns either True or False

```python
x = 10
y = 5
z = 10
# Let us come up with all True statements using various comparison operators
print(x > y) # x is greater than y
print(y < x) # y is less than x
print(x != y) # x is not equals to y
print(x == z) # x is equals to z
print(x >= z) # x is greater than or equals to z
print(x <= z) # x is less than or equals to z
```

# Logical Operators
Compare more than one condition and always return either True or False

```python
# I want to book a hotel room. My budget is RM150 and below. However I want atleast a 3star hotel.
# In this case I will go to booking.com and search for a hotel room that satisfy both conditions.
# We want both condition to be True, then we use "and" operator
budget = 150
requiredstar = 3
roomprice = 140
hotelstar = 4
print(roomprice <= budget and hotelstar >= requiredstar)
# Truth table for and operator
# True   and  True   =  True
# True   and  False  =  False
# False  and  True   =  False
# False  and  False  =  False
```

```python
# I want to buy a Television. My budget is 1500.
# However if the brand is "Sony" I can go more than the budget.

# Truth table for or operator
# True   or  True   =  True
# True   or  False  =  True
# False  or  True   =  True
# False  or  False  =  False

mybudget = 1500
preferredmodel = "Sony"
tvprice = 1550
tvmodel = "Sony"
print(tvprice <= mybudget or tvmodel == preferredmodel)
```
