# Tuple

Tuple is nothing but readonly list

```python
# Instead of using square bracket we use round bracket () to create a tuple
# Remember tuple is nothing but readonly list
# Meaning selection and indexing will be exactly same as list
# However methods like append, remove, pop are not available in the list
countries = ("Indonesia", "Malaysia", "Singapore", "Thailand")
print(countries)
print(countries[0])
print(countries[1])
print(countries[2])
print(countries[3])
print(len(countries))
print(countries[-1])
print(countries[-2])
print(countries[-3])
# countries.append("Vietnam") # this will throw an error 'tuple' object has no attribute 'append'
# Remember it is readonly list
```

```python
# Since tuple is readonly, performance wise it is faster than a list
# Tuple is a preferred data structure for the python language.
prices = 10.5, 20.5, 30.5, 40.5, 50.5 # Here we did not use () or [] Values are just seperated by comma
# and assigned to the variable prices.
# Python will automatically place the values inside the tuple and assign the tuple to the variable prices
print(prices, type(prices))
```

The () is used for the following purposes
1. call a function ex. print()
2. evaluate expression ex. (10 + 5) * 3

```python
# I want to book a hotel room. My budget is RM150 and below. However I want atleast a 3star hotel.
# In this case I will go to booking.com and search for a hotel room that satisfy both conditions.
# We want both condition to be True, then we use "and" operator
budget = 150
requiredstar = 3
roomprice = 140
hotelstar = 4
print(roomprice <= budget and hotelstar >= requiredstar)
# Truth table for and operator
# True   and  True   =  True
# True   and  False  =  False
# False  and  True   =  False
# False  and  False  =  False
```

```python
# I want to buy a Television. My budget is 1500.
# However if the brand is "Sony" I can go more than the budget.

# Truth table for or operator
# True   or  True   =  True
# True   or  False  =  True
# False  or  True   =  True
# False  or  False  =  False

mybudget = 1500
preferredmodel = "Sony"
tvprice = 1550
tvmodel = "Sony"
print(tvprice <= mybudget or tvmodel == preferredmodel)
```

```python
# We thought the control will come out of the loop only after iterating all the items inside the list
# However in python we have a keyword break that can be use to come out of the loop without iterating
# all the items inside the list

# Find whether durian is inside the list
fruits = ["apple", "orange", "mango", "durian", "banana", "grapes"]
for fruit in fruits:
  if fruit == "durian":
    print("Durian is inside the list")
    break
else:
  # the code in the else block will not get executed because the break keyword gets the control
  # out of the loop the moment it sees durian
  # In other words not all fruits are iterated
  print("Durian is not inside the list")
```

# While Loop

We do not have list of items and we do not know how many times the block of code has to be executed.


1. Get password as user input, validate the keyed in password against the password in database
2. If they are not same then ask the user to key in the password again repeat this process until the user key in the correct password.

1. Is there a list ? - No
2. Do we know how many time the user is going to make mistake ? - No


This is exactly where we use while loop

```python
passwordfromdb = "Password"
# the following is an infinite loop
# Usually we will say while (condition)
# However in this case we hardcoded the condition as True (always True)
# It will never become False that is why it is an infinite loop
# But we already know to come out of a loop you can use the keyword break
while True:
  password = input("Key in the password: ")
  if password == passwordfromdb:
    print("Password is correct")
    break
  else:
    print("Password is incorrect")
```

```python
def simpleInterest(principle, period, rate):
  interest = (principle * period * rate) / 100
  return interest
```

# Special features of Function

## Optional Parameters

```python
# We know the arguments are positional
# However in python we can create function with optional parameters
# I want to make period and rate as optional parameter
# You can do that by setting default values to those parameter
# Principle is a compulsory parameter
def simpleInterest(principle, period=1, rate=6):
  interest = (principle * period * rate) / 100
  return interest
```

```python
# It will take 1 for period and 6 for rate and interest is calculated
print(simpleInterest(1000))

# Since we pass 2 which will overwrite the default value of period which is 1
# It will take 6 for rate and interest is calculated
print(simpleInterest(1000, 2))

print(simpleInterest(1000, 2, 5))
```

```python
# def simpleInterest(principle, period=1, rate=6):
# We know arguments are positional. Now can we pass argument for principle and rate only
# In python they have another concept called keyword argument
simpleInterest(principle=1000, rate=5)
# Here the keyword is nothing but the parameter name
```

## Variable Length Arguments

```python
# To handle variable length arguments in python you can use *parameter
# Since we have * in the parameter, python will automatically take all the arguments
# place them inside a list and send the list as argument
def calculateTotal(*mynumbers):
  total = 0
  for number in mynumbers:
    total = total + number
  return total
```

```python
# This time I am not going to place the values inside the list
# I am going to send the values individually
print(calculateTotal(10, 20, 30, 40, 50, 60))
print(calculateTotal(10, 20, 30, 40, 50, 60, 70, 80, 90))
print(calculateTotal(10, 20, 30))
```

```python
mynumbers = [10, 20, 30, 40, 50, 60]
# Since the parameter has * python will take this list and place it inside another list
# it becomes a 2 dimensional list, howeve the function is looking for single dimensional list
# print(calculateTotal(mynumbers))
# we must unpack the items inside the list and send them as individual values
print(calculateTotal(*mynumbers))
# When the argument has *, then the values inside the list (argument) is automatically unpacked
# to individual values.
# Remember the function parameter already has *, so python will pack the
# individual values to a list and pass it as a single dimensional list.
# Note:
# When you have * in the argument python will unpack the list of items to individual items
# When you have * in the parameter python will pack the values to a list
```

# Runtime Exception Handling

```python
x = input("Give me a number: ")
y = input("Give me another number: ")
# the error happens because of the input
# this type of error is called run time errors
# Usually if there is a statement related to I/O
# there is a high chance of that statement to throw error in future
# We must handle this kind of runtime exception and in python we have
# try except else finally block
try:
  # the code that might throw error in future should go inside here
  x = int(x)
  y = int(y)
except:
  # during runtime if any error occur because of input
  # the code to handle that error should be placed inside except
  print("Invalid Input")
else:
  # during runtime if there is no error occur because of input
  # the code to be executed when no error occur should be placed inside else
  print(x + y)
finally:
  # the code regardless of whether there is runtime error or no error
  # gets executed should be place inside the finally block
  print("Thank You")
```
